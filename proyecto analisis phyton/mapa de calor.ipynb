```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import HeatMap

# Cargar datos de ingresantes y datos de coordenadas
file_path_ingresantes = 'DATA_INGRESANTES_SIN_ACENTOS.csv'
file_path_coordenadas = 'distritos_coordenadas_actualizado.csv'

df_ingresantes = pd.read_csv(file_path_ingresantes)
df_coordenadas = pd.read_csv(file_path_coordenadas)

# Obtener distritos únicos de procedencia
distritos_unicos = df_ingresantes['DISTRITO_NACIMIENTO'].unique()

# Filtrar coordenadas para los distritos de los ingresantes
df_distritos_ingresantes = df_coordenadas[df_coordenadas['distrito'].isin(distritos_unicos)]

# Crear un DataFrame de datos para el mapa de calor
heat_data = []
for _, row in df_distritos_ingresantes.iterrows():
    distrito = row['distrito']
    lat = row['lat']
    lon = row['lon']
    count = df_ingresantes[df_ingresantes['DISTRITO_NACIMIENTO'] == distrito].shape[0]
    heat_data.append([lat, lon, count])

# Crear un mapa centrado en Perú
m = folium.Map(location=[-9.19, -75.0152], zoom_start=6)


# Agregar mapa de calor
HeatMap(heat_data).add_to(m)

# Guardar el mapa en un archivo HTML
m.save('mapa_de_calor_personalizado.html')

# Mostrar el mapa
m


```
